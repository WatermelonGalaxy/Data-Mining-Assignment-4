```python
# import libraries
import pandas as pd
import numpy as np
import mlxtend as ml
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# import emoji dataset
emoji = pd.read_csv('emoji-mainia.csv', header=None)
emoji.head()

# convert dataset to list
emoji_list = emoji.values.tolist()
#emoji_list = np.array(emoji_list)

# drop nan values
for i in range(len(emoji_list)):
    emoji_list[i] = [x for x in emoji_list[i] if pd.notnull(x)]

emoji_list
```

```python
# creat the encoded dataset
te = TransactionEncoder()
te_ary = te.fit(emoji_list).transform(emoji_list)

# convert the encoded dataset into a dataframe
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
# Apriori Algorithm function from mlxtend
apriori(df, min_support=.375, use_colnames=True)
```
